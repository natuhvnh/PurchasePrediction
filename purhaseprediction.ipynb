```python
train = pd.read_csv('train.csv')
train
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(train, test_size=0.1, random_state=42, shuffle=True)
```

```python
train
```

```python
test
```

```python
# task(label='label', path='model').fit(train, auto_stack=True, presets='best_quality')
task(label='label', path='model').fit(train, num_bag_folds=5, num_bag_sets=1, num_stack_levels=2)
# task(label='label', path='model').fit(train, auto_stack=True)
```

```python
y_test = test['label']
test_data = test.drop(['label'], axis=1)
test_data
```

```python
predictor = task.load('model')
y_pred = predictor.predict(test_data)
# print("Predictions:  \n", y_pred)
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```
